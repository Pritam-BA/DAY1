```python
# 1. Exploring customer data using dictionary operations
print("=== CUSTOMER DATA EXPLORATION ===")

# Get all customer IDs (dictionary keys)
customer_ids = list(customers_data.keys())
print(f"Customer IDs: {customer_ids}")

# Access specific customer information
print(f"\nCustomer CUST001 Details:")
for key, value in customers_data['CUST001'].items():
    print(f"  {key}: {value}")

# Find customers by criteria using dictionary comprehension
high_balance_customers = {
    cust_id: details for cust_id, details in customers_data.items()
    if details['balance'] > 80000
}
print(f"\nHigh Balance Customers (>80K): {list(high_balance_customers.keys())}")
```

```python
# 2. Transaction analysis using list operations
print("\n=== TRANSACTION ANALYSIS ===")

# Filter transactions by type using list comprehension
credit_transactions = [txn for txn in transactions_data if txn['type'] == 'Credit']
debit_transactions = [txn for txn in transactions_data if txn['type'] == 'Debit']

print(f"Total Credit Transactions: {len(credit_transactions)}")
print(f"Total Debit Transactions: {len(debit_transactions)}")

# Calculate total amounts
total_credits = sum([txn['amount'] for txn in credit_transactions])
total_debits = sum([txn['amount'] for txn in debit_transactions])

print(f"Total Credit Amount: ₹{total_credits:,}")
print(f"Total Debit Amount: ₹{total_debits:,}")
```

```python
# 3. Create customer-wise transaction summary
print("\n=== CUSTOMER TRANSACTION SUMMARY ===")

# Initialize customer transaction summary dictionary
customer_summary = {}

# Process each transaction
for txn in transactions_data:
    cust_id = txn['customer_id']

    # Initialize customer entry if not exists
    if cust_id not in customer_summary:
        customer_summary[cust_id] = {
            'total_credits': 0,
            'total_debits': 0,
            'transaction_count': 0,
            'transactions': []
        }

    # Update summary
    customer_summary[cust_id]['transaction_count'] += 1
    customer_summary[cust_id]['transactions'].append(txn)

    if txn['type'] == 'Credit':
        customer_summary[cust_id]['total_credits'] += txn['amount']
    else:
        customer_summary[cust_id]['total_debits'] += txn['amount']

# Display summary
for cust_id, summary in customer_summary.items():
    customer_name = customers_data[cust_id]['name']
    net_flow = summary['total_credits'] - summary['total_debits']
    print(f"{customer_name} ({cust_id}):")
    print(f"  Credits: ₹{summary['total_credits']:,}")
    print(f"  Debits: ₹{summary['total_debits']:,}")
    print(f"  Net Flow: ₹{net_flow:,}")
    print(f"  Transactions: {summary['transaction_count']}")
    print()
```

```python
# 4. Convert to pandas DataFrame for advanced analysis
print("=== CONVERTING TO DATAFRAMES ===")

# Convert customer data to DataFrame
customer_df = pd.DataFrame.from_dict(customers_data, orient='index')
customer_df.reset_index(inplace=True)
customer_df.rename(columns={'index': 'customer_id'}, inplace=True)

# Convert transaction data to DataFrame
transaction_df = pd.DataFrame(transactions_data)
transaction_df['date'] = pd.to_datetime(transaction_df['date'])
transaction_df['amount'] = pd.to_numeric(transaction_df['amount'])

print("Customer DataFrame:")
print(customer_df.head())
print(f"\nTransaction DataFrame:")
print(transaction_df.head())
```

```python
# 5. Define business analysis functions
def get_customer_profile(customer_id):
    """Get complete customer profile with transactions"""
    if customer_id not in customers_data:
        return "Customer not found"

    profile = customers_data[customer_id].copy()
    profile['transactions'] = [txn for txn in transactions_data if txn['customer_id'] == customer_id]
    profile['total_transactions'] = len(profile['transactions'])

    return profile

def calculate_customer_metrics():
    """Calculate key business metrics for all customers"""
    metrics = {}

    for cust_id in customers_data.keys():
        customer_txns = [txn for txn in transactions_data if txn['customer_id'] == cust_id]

        if customer_txns:
            amounts = [txn['amount'] for txn in customer_txns]
            metrics[cust_id] = {
                'avg_transaction_amount': np.mean(amounts),
                'max_transaction': max(amounts),
                'min_transaction': min(amounts),
                'transaction_frequency': len(customer_txns),
                'customer_name': customers_data[cust_id]['name']
            }

    return metrics

# Execute analysis
customer_metrics = calculate_customer_metrics()
print("Customer Metrics:")
for cust_id, metrics in customer_metrics.items():
    print(f"{metrics['customer_name']}: Avg Txn ₹{metrics['avg_transaction_amount']:,.0f}")
```

```python
# 6. Create business visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Banking Customer Data Analysis Dashboard', fontsize=16, fontweight='bold')

# 1. Customer Balance Distribution
axes[0,0].bar(customer_df['name'], customer_df['balance'], color='lightblue', edgecolor='navy')
axes[0,0].set_title('Customer Account Balances')
axes[0,0].set_xlabel('Customer Name')
axes[0,0].set_ylabel('Balance (₹)')
axes[0,0].tick_params(axis='x', rotation=45)

# 2. Credit Score vs Balance
axes[0,1].scatter(customer_df['credit_score'], customer_df['balance'],
                  s=100, c=customer_df['age'], cmap='viridis', alpha=0.7)
axes[0,1].set_title('Credit Score vs Account Balance')
axes[0,1].set_xlabel('Credit Score')
axes[0,1].set_ylabel('Balance (₹)')
colorbar = plt.colorbar(axes[0,1].collections[0], ax=axes[0,1])
colorbar.set_label('Age')

# 3. Transaction Type Distribution
transaction_summary = transaction_df.groupby('type')['amount'].sum()
axes[1,0].pie(transaction_summary.values, labels=transaction_summary.index,
              autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral'])
axes[1,0].set_title('Transaction Amount Distribution')

# 4. State-wise Customer Distribution
state_counts = customer_df['state'].value_counts()
axes[1,1].bar(state_counts.index, state_counts.values, color='orange', alpha=0.7)
axes[1,1].set_title('Customers by State')
axes[1,1].set_xlabel('State')
axes[1,1].set_ylabel('Number of Customers')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Additional Analysis: Transaction Timeline
plt.figure(figsize=(12, 6))
daily_transactions = transaction_df.groupby(['date', 'type'])['amount'].sum().unstack(fill_value=0)
daily_transactions.plot(kind='bar', stacked=True, color=['lightgreen', 'lightcoral'])
plt.title('Daily Transaction Amounts by Type')
plt.xlabel('Date')
plt.ylabel('Amount (₹)')
plt.legend(['Credit', 'Debit'])
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
